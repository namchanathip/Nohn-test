
````python
x = 10
y = 12
if x > y:
    print("x>y")
elif x < y:
    print("x<y")
else:
    print("x=y")
````

### Break

As the name implies, it is used to break a loop when the condition becomes true during loop execution.

````python
for i in range(100):
    if i>=7:
        break
    print(i)
````

### Continue

`continue` continues with the next iteration of the loop.

````python
for val in "string":
    if val == "i":
        continue
    print(val)

print("The end")
````
